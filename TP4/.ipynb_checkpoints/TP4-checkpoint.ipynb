```python
import numpy as np
import skimage as ski
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

print(ski.__version__)
```

```python
img = ski.io.imread("cgefug8s28881.jpg", as_gray=True)

print(img.shape[:2])
print(img.shape)
print(img.dtype)
```

```python
# Afficher l'image d'entrée
def showImg(img, title):
    plt.figure(figsize=(10, 5))
    plt.subplot(2, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')

showImg(img, "Image d'entrée")
```

```python
canny = ski.feature.canny(img, sigma=2, low_threshold=0.8, high_threshold=0.95)

showImg(canny, "Canny")
```

```python
def gaussianFiltering(f, s):
    return ski.filters.gaussian(f, sigma=s, output=None, mode='nearest', cval=0, preserve_range=False, truncate=4.0)

gauss = gaussianFiltering(img, 5)
showImg(gauss, "Gauss")
```

```python
def computeGx(f):
    S_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
    return convolve2d(f, S_x, mode='same', boundary='symm')

def computeGy(f):
    S_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])
    return convolve2d(f, S_y, mode='same', boundary='symm')

def computeMagnitude(Gx, Gy):
    return np.sqrt(Gx**2 + Gy**2)

def computeDirection(Gx, Gy):
    return np.arctan2(Gy, Gx)
```

```python
gx = computeGx(img)
gy = computeGy(img)
mag = computeMagnitude(gx, gy)
dir = computeDirection(gx, gy)
```

```python
showImg(mag, "Mag")
showImg(dir, "Dir")
```

```python
def removeNonMaxima(grad_m, grad_d):
    grad_m_max = np.zeros(grad_m.shape)
    rows, cols = grad_m.shape
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            angle = grad_d[i, j]
            if angle < -np.pi / 8:
                angle += np.pi
            elif angle >= 7 * np.pi / 8:
                angle -= np.pi
            
            if -np.pi / 8 <= angle < np.pi / 8 or 7 * np.pi / 8 <= angle:
                neighbors = [grad_m[i, j-1], grad_m[i, j+1]]
            elif np.pi / 8 <= angle < 3 * np.pi / 8:
                neighbors = [grad_m[i-1, j+1], grad_m[i+1, j-1]]
            elif 3 * np.pi / 8 <= angle < 5 * np.pi / 8:
                neighbors = [grad_m[i-1, j], grad_m[i+1, j]]
            else:
                neighbors = [grad_m[i-1, j-1], grad_m[i+1, j+1]]
            
            if grad_m[i, j] >= max(neighbors):
                grad_m_max[i, j] = grad_m[i, j]
    
    return grad_m_max
```

```python
non_max = removeNonMaxima(mag, dir)

showImg(non_max, "Non_Max")
```

```python
def computeThresholds(grad_maxima, alpha, beta):
    # Flatten the gradient maxima image
    grad_maxima_flat = grad_maxima.flatten()
    
    # Sort the values in ascending order
    sorted_values = np.sort(grad_maxima_flat)
    
    # Compute the high threshold
    t_high_index = round(alpha * len(sorted_values))
    t_high = sorted_values[t_high_index]
    
    # Compute the low threshold
    t_low = beta * t_high
    
    return t_high, t_low

t_high, t_low = computeThresholds(non_max, 0.8, 0.95)

print(t_high, t_low)
```

```python
from queue import Queue

def hysteresisThresholding(grad_maxima, tLow, tHigh):
    rows, cols = grad_maxima.shape
    canny = np.zeros_like(grad_maxima)
    fifo = Queue()
    
    # Initial pass to identify strong edges and populate FIFO queue
    for i in range(rows):
        for j in range(cols):
            if grad_maxima[i, j] >= tHigh:
                fifo.put((i, j))
                canny[i, j] = 255
            elif grad_maxima[i, j] < tLow:
                canny[i, j] = 0
    
    # Hysteresis thresholding
    while not fifo.empty():
        p = fifo.get()
        p_row, p_col = p[0], p[1]
        neighbors = [(p_row-1, p_col-1), (p_row-1, p_col), (p_row-1, p_col+1),
                     (p_row, p_col-1),                     (p_row, p_col+1),
                     (p_row+1, p_col-1), (p_row+1, p_col), (p_row+1, p_col+1)]
        
        for neighbor in neighbors:
            n_row, n_col = neighbor[0], neighbor[1]
            if 0 <= n_row < rows and 0 <= n_col < cols and grad_maxima[n_row, n_col] >= tLow and canny[n_row, n_col] == 0:
                canny[n_row, n_col] = 255
                fifo.put((n_row, n_col))
    
    return canny
```

```python
hyst = hysteresisThresholding(non_max, t_low, t_high)

showImg(hyst, "Hyst")
```
