```python
import numpy as np
import skimage as ski
import matplotlib.pyplot as plt

img = ski.io.imread("cgefug8s28881.jpg", as_gray=True)

# Afficher l'image d'entrée
def showImg(img, title):
    plt.figure(figsize=(20, 10))
    plt.subplot(2, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')

showImg(img, "Image d'entrée")
```

```python
square_footprint = ski.morphology.square(3);
disk_footprint = ski.morphology.disk(7)
```

```python
erosion = ski.morphology.erosion(img, square_footprint)

showImg(erosion, "Erosion");
```

```python
dilation = ski.morphology.dilation(img, square_footprint)

showImg(dilation, "Dilatation")
```

```python
opening = ski.morphology.opening(img, square_footprint)

showImg(opening, "Ouverture")
```

```python
closing = ski.morphology.closing(img, square_footprint)

showImg(closing, "Fermeture")
```

```python
erosion = ski.morphology.erosion(img, disk_footprint)

showImg(erosion, "Erosion")
```

```python
dilation = ski.morphology.dilation(img, disk_footprint)

showImg(dilation, "Dilatation")
```

```python
opening = ski.morphology.opening(img, disk_footprint)

showImg(opening, "Ouveture")
```

```python
closing = ski.morphology.closing(img, disk_footprint)

showImg(closing, "Closing")
```

```python
img = ski.io.imread("figs/lines.png", as_gray=True)

showImg(img, "Image d'entrée")
```

```python
footprint = ski.transform.rotate(np.ones((1, 20), dtype="uint8"), -45, resize=True, order=0)

closing = ski.morphology.closing(img, footprint)

showImg(closing, "Closing")
```

```python
img = ski.io.imread("figs/rayures.jpg", as_gray=True)

showImg(img, "Image d'entrée")
```

```python
footprint = np.ones((1, 8), dtype="uint8")

closing = ski.morphology.closing(img, footprint)

showImg(closing, "Closing")
```

```python
img = ski.io.imread("figs/circuit.png", as_gray=True)

showImg(img, "Image d'entrée")
```

```python
g = (img<80).astype('uint8')

showImg(g, "Seuil")
```

```python
closing = ski.morphology.closing(g, ski.morphology.disk(2))

showImg(closing, "Fermeture")
```

```python
opening = ski.morphology.opening(closing, ski.morphology.disk(8))

showImg(opening, "Ouverture")
```

```python
diff = (opening - closing).astype('uint8')

showImg(diff, "Diff")
```

```python
opening = ski.morphology.opening(diff, ski.morphology.disk(1))

showImg(opening, "Ouverture")
```
